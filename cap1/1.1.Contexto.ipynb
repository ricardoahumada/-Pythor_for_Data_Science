### Aprendizaje supervisado vs Aprendizaje no supervisado


La principal diferencia es la presencia (o no) de columna respuesta (explicada, objetivo, dependiente...). 

El **aprendizaje supervisado** intenta predecir el valor de esa columna (si el usuario comprará un producto, la esperanza de vida, el riesgo de morosidad...) usando otras columnas de datos.

El **aprendizaje no supervisado** solamente usa esas otras columnas para organizar los datos, agrupándolos en categorías que previamente no existían.

# Desafíos del ML

- **Calidad de los datos:** tener datos de buena calidad para los algoritmos de ML es uno de los mayores desafíos. El uso de datos de baja calidad genera problemas relacionados con el preprocesamiento de datos y la extracción de características.

- **Tareas que consumen mucho tiempo:** otro desafío que enfrentan los modelos ML es el consumo de tiempo, especialmente para la adquisición de datos, la extracción y recuperación de características.

- **Falta de especialistas:** dado que la tecnología ML aún está en su etapa inicial, la disponibilidad de recursos expertos es un trabajo difícil.

- **No objetivos claros para formular los problemas de negocio:** no tener un objetivo claro y una meta bien definida para los problemas de negocio es otro desafío clave.

- **Problema de sobreajuste (overfitting) y desajuste(underfitting):** si el modelo se sobreajusta o no se ajusta bien, no se ajustará bien para el problema.

- **La maldición de la dimensionalidad:** el exceso de variables (features) de los puntos de datos. Esto puede ser un verdadero obstáculo.

- **Dificultad en la implementación:** la complejidad del modelo ML hace que sea bastante difícil implementarlo en la vida real.
