## Algoritmos de clasificacin

>Logistic Regression, Naive Bayes, Support Vector Machine, Decision Tree, Random Forest, Gradient Boosting Machine (GBM), Perceptron, Linear Discriminant Analysis (LDA), K Nearest Neighbours (KNN), Learning Vector Quantisation (LVQ) or Neural Network.

**Factores a considerar:** 
- linealidad: datos separados por una lnea recta
- interpretabilidad: la capacidad de explicar por qu un punto de datos se clasifica en una determinada clase
- multiclase: dos clases (clasificacin binaria) o ms de dos clases (multiclase). Support Vector Machine (SVM), Linear Discriminant Analysis (LDA) y Perceptron son clasificacin binaria, pero todo lo dems se puede usar tanto para binario como para multiclase.



#### 1. Regresin logstica

**Fortalezas:** buena precisin en pequeas cantidades de datos, fcil de interpretar (obtenemos la importancia de las features), fcil de implementar, eficiente para entrenar (no necesita una gran potencia de cmputo), puede hacer multiclase.

**Debilidades:** tiende a sobreajustarse en dimensiones altas (usar regularizacin), no puede hacer una clasificacin no lineal (o una relacin compleja), no es bueno con la multicolinealidad, es sensible a los valores atpicos, requiere una relacin lineal entre las probabilidades logartmicas y la variable objetivo.

#### 2. Naive Bayes

**Fortalezas:** buena precisin en pequeas cantidades de datos, eficiente para entrenar (no necesita una gran potencia de cmputo), fcil de implementar, altamente escalable, puede hacer multiclase, puede hacer datos continuos y discretos, no es sensible a features irrelevantes.

**Debilidades:** las features deben ser independientes, una categora que existe en el conjunto de datos de prueba pero no en el conjunto de datos de entrenamiento tendr probabilidad cero (problema de frecuencia cero)

#### 3. Random Forest

***Fortalezas:** fcil de interpretar (intuitivo, muestra interaccin entre variables), puede clasificar datos no lineales, no es necesario normalizar ni escalar los datos, no se ve afectado por valores faltantes, no se ve afectado por valores atpicos, funciona bien con datos desequilibrados (el la naturaleza de la distribucin de datos no importa), puede hacer clasificacin y regresin, puede hacer datos numricos y categricos, proporciona importancia de features (calculada a partir de la disminucin en la impureza del nodo), bueno con un gran conjunto de datos, capaz de manejar multicolinealidad.

**Debilidades:** tiene tendencia a sobreajustarse (sesgo hacia el conjunto de entrenamiento, requiere poda), no es robusto (alta varianza, un pequeo cambio en los datos de entrenamiento da como resultado un cambio importante en el modelo y la salida), no es bueno con la variable continua, requiere ms tiempo para entrenar el modelo (muchos recursos).

#### 4. Random Forest

**Fortalezas:** alta precisin, no necesita poda, sin sobreajuste, bajo sesgo con una varianza bastante baja/moderada (debido al arranque), puede hacer tanto clasificacin como regresin, puede hacer numrico y categrico, puede clasificar datos no lineales, los datos no no es necesario normalizarlo ni escalarlo, no se ve afectado por valores faltantes, no se ve afectado por valores atpicos, funciona bien con datos desequilibrados (la naturaleza de la distribucin de datos no importa), se puede paralelizar (puede usar mltiples CPU en paralelo), bueno con alta dimensionalidad.

**Debilidades:** largo tiempo de entrenamiento, requiere mucha memoria, no interpretable (porque hay cientos de rboles).

#### 5. Support Vector Machine (lineal)

**Fortalezas:** se escala bien con datos dimensionales altos, estable (baja varianza), menos riesgo de sobrevaloracin, no depende de todos los datos (no se ve afectado por los valores faltantes), funciona bien con ruido.

**Debilidades:** largo tiempo de entrenamiento para grandes datos, requiere escalado de features.

#### 6. Support Vector Machine (con Kernel)

**Fortalezas:** escala bien con datos dimensionales altos, estable (baja varianza), maneja muy bien los datos no lineales, menos riesgo de sobrevaloracin (debido a la regularizacin), bueno con valores atpicos (tiene gamma y C para controlar), puede detectar valores atpicos en anomalas deteccin, funciona bien con ruido.

**Debilidades:** largo tiempo de entrenamiento para grandes datos, difcil encontrar el kernel apropiado, necesita mucha memoria, requiere escalado de funciones, difcil de interpretar.

#### 7. Gradient Boosting

**Fortalezas:** alta precisin, flexible con varias funciones de prdida, preprocesamiento mnimo, no se ve afectado por los valores faltantes, funciona bien con datos desequilibrados, puede hacer clasificacin y regresin.

**Debilidades:** tendencia a sobreajustarse (porque contina minimizando errores), sensible con valores atpicos, gran requerimiento de memoria (miles de rboles), largo tiempo de entrenamiento, bsqueda de hiperparmetros en grillas grandes, no es bueno con ruido, difcil de interpretar.

#### 7. K Nearest Neighbours

**Fortalezas:** simple de entender (intuitivo), simple de implementar (tanto binario como multiclase), maneja bien los datos no lineales, no paramtrico (sin requisitos en la distribucin de datos), responde rpidamente a los cambios de datos en la implementacin en tiempo real, puede hacer tanto de clasificacin como de regresin.

**Debilidades:** largo tiempo de entrenamiento, no funciona bien con datos de alta dimensin, requiere escalado, no funciona bien con datos desequilibrados, sensible a valores atpicos y ruido, afectado por valores faltantes.

#### 8. Red neuronal

**Fortalezas:** alta precisin, maneja bien datos no lineales, generaliza bien en datos no vistos (baja varianza), no paramtrico (sin requisitos en la distribucin de datos), trabaja con datos heterocedsticos (varianza no constante), trabaja con datos altamente voltiles (tiempo serie), funciona con datos incompletos (no afectados por valores faltantes), tolerancia a fallas

**Debilidades:** requiere una gran cantidad de datos, computacionalmente costoso (requiere procesadores paralelos/GPU y gran memoria), no interpretable, complicado para obtener la arquitectura correcta (#capas, #neuronas, funciones, etc.)
